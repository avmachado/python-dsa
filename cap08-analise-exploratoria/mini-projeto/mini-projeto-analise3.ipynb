### Análise 3

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.columns
```

#### Calcule a média de preço por marca e por veículo

```python
df2 = df.groupby(['brand', 'vehicleType']).mean()
df2 = pd.DataFrame(df2['price'])
df2 = df2.reset_index() # Função usada para transformar os índices em colunas
df2.head()
```

### Preço médio de um veículo por marca, bem como tipo de veículo

```python
df3 = df2.pivot('brand', 'vehicleType', 'price')

f2, ax = plt.subplots(figsize = (12, 20))

# sns.set_theme()

ax = sns.heatmap(df3)
plt.xlabel('Tipo do Veículo', fontsize = 16)
plt.ylabel('Marca', fontsize = 16)
plt.title('Preço Médio por Marca e Tipo do Veículo', fontsize = 20)
```
