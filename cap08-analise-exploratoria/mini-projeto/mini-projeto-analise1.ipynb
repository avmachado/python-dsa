## Análise Exploratória em Conjunto de Dados do Kaggle

### Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Importação do Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

### Distribuição de veículos com base no ano de registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
```

```python
df.head()
```

```python
df.columns
```

```python
# Construindo o gráfico
f, ax = plt.subplots(figsize=(10, 8))
sns.distplot(df.yearOfRegistration, kde = True)
plt.xlabel('Ano de Registro', fontsize = 18)
plt.ylabel('Densidade', fontsize = 18)
plt.title('Distribuição de Veículos por Ano de Registro', fontsize = 20)
```

```python
# O ano de registro dos veículos se concentra entre os anos de 1990 e 2010, aproximadamente.
# Salvando o plot
f.savefig("plots/analise1/vehicle-distribution.png")
```

### Variação de Preço pelo Tipo do Veículo

```python
# Crie um Boxplot para avaliar os outliers
```

```python
# Construindo o gráfico
f2, ax2 = plt.subplots(figsize=(10, 8))
sns.boxplot(data=df, x = 'vehicleType', y = 'price')
plt.xlabel('Tipo do Veículo', fontsize = 18)
plt.ylabel('Preço', fontsize = 18)
plt.title('Tipo do Veículo x Preço', fontsize = 20)
plt.show()
```

```python
# Salvando o plot
f2.savefig("plots/analise1/price-vehicleType-boxplot.png")
```

### Contagem total de veículos à venda conforme o tipo do veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria
```

```python
# Construindo o gráfico
f3 = plt.figure()

ax = sns.countplot(data=df, x = 'vehicleType')
sns.set(rc={'figure.figsize':(11,8)})
ax.set_xlabel('Tipo do Veículo', fontsize = 16)
ax.set_ylabel('Contagem dos Veículos', fontsize = 16)
ax.set_title('Contagem de Veículos por Tipo', fontsize = 20)
```

```python
# Salvando o plot
f3.savefig("plots/analise1/count-vehicleType.png")
```
