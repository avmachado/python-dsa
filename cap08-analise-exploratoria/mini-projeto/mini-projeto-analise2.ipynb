```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.columns
```

### Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
```

```python
# Definindo os dados
num_veiculos = df['brand'].groupby(df.brand).count().sort_values()

# Criando o gráfico
f1, ax1 = plt.subplots(figsize = (12, 8))

ax1.barh(num_veiculos.index, num_veiculos.values)
plt.yticks(num_veiculos.index)
plt.xlabel('Número de Veículos', fontsize = 16)
plt.ylabel('Marca', fontsize = 16)
plt.title('Número de Veículos por Marca')
plt.show()
```

```python
# Salvando o plot
f1.savefig(("plots/analise2/brand-vehicleCount.png"))
```

### Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
```

```python
# Criando o gráfico
f2, ax2 = plt.subplots(figsize = (12, 8))

sns.barplot(x="vehicleType", y="price",hue="gearbox", data=df)
plt.xlabel('Tipo do Veículo', fontsize = 16)
plt.ylabel('Preço Médio', fontsize = 16)
plt.title('Preço Médio por Tipo de Veículo e de Caixa de Câmbio')
```

```python
# Salvando o plot
f2.savefig("plots/analise2/vehicletype-gearbox-price.png")
```
